```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_histogram(column_df, column_name):
    plt.figure(figsize=(5, 3))
    plt.title(f"Distribution of {column_name}")
    sns.histplot(column_df, kde=True) 
    plt.show()
```

```python
plot_histogram(df['type'], 'Type')
```

```python
plot_histogram(df['mfr'], 'Manufacturer')
```

```python
plot_histogram(df['calories'], 'Calorie')
```

```python
plt.figure(figsize=(10, 6))
plt.title("Brands and Their Ratings")
plt.xticks(rotation=90)
sns.barplot(data=df, x='name', y='rating')
plt.show()
```

```python
correlation = df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title(' Correlation Between Nutrients')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title("Box Plot of Data")
plt.show()
```

```python
sns.pairplot(df)
plt.title("Pair Plot of Data")
plt.show()
```
